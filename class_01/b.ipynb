### Ruina gracza dla 2 graczy A i B

Zał:
- a + b = 100
- pA = 50%

Szukane:

Prawdopodobieństwo ruiny gracza A w zależnosci od salda początkowego gracza A

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def coin_game(a):
    b = 100 - a
    while a > 0 and b > 0:
      roll = np.random.uniform(0, 100)
      if 50 > roll:
        b += 1
        a -= 1
      else:
        a += 1
        b -= 1
    return b > 0
```

```python
simulations = 500
a = np.arange(1, 101, 1)
results = {}
```

```python
for value in a:
  count = 0

  for _ in range(simulations):
    if coin_game(value): count += 1

  results[value] = count / simulations
```

```python
theoretical_probability = []
for i in results.keys():
    theoretical_probability.append(1 - i / 100)
```

```python
x_values = list(results.keys())
y_values = list(results.values())

plt.figure(figsize=(10, 6))
plt.scatter(x_values, y_values, color='blue', alpha=0.7)
plt.plot(x_values, theoretical_probability, color='grey', linestyle='--')
plt.title('Prawdopodobieństwo ruiny gracza A w zależności od salda początkowego gracza A')
plt.xlabel('Saldo poczatkowe A')
plt.ylabel('Prawdopodobieństwo ruiny gracza A')
plt.grid(True)
plt.show()
```
